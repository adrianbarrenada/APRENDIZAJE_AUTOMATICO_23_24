Antes de querer plantear ningún modelo SARIMA para poder predecir los datos a futuro, nuestros datos deben ser estacionarios. Vamos a comenzar por visualizar el ACF y el PACF de nuestra base de datos mensual.

```python
monthly_df
```

```python
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf


# Ordenar la base de datos por año y mes
monthly_df = monthly_df.sort_values(by=['YEAR', 'MONTH'])

# Graficar ACF y PACF
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# ACF
plot_acf(monthly_df['TOTAL MP'], lags=15, ax=axes[0])
axes[0].set_title("Autocorrelation Function (ACF) de TOTAL MP mensualmente")

# PACF
plot_pacf(monthly_df['TOTAL MP'], lags=15, ax=axes[1])
axes[1].set_title("Partial Autocorrelation Function (PACF) de TOTAL MP mensualmente")

plt.tight_layout()
plt.show()
```

Vemos que los datos no parecen estacionarios estacionarios. Vamos a comprobarlo.

```python
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Cargar los datos
# Asegúrate de que 'df' es el DataFrame con la columna 'TOTAL MP'
df = monthly_df[['YEAR', 'MONTH', 'TOTAL MP']]
df['DATE'] = pd.to_datetime(df[['YEAR', 'MONTH']].assign(day=1))
df.set_index('DATE', inplace=True)
df = df[['TOTAL MP']]

def check_stationarity(series):
    result = adfuller(series)
    print("ADF Statistic:", result[0])
    print("p-value:", result[1])
    if result[1] > 0.05:
        print("La serie NO es estacionaria")
    else:
        print("La serie es estacionaria")

# Verificar estacionariedad
check_stationarity(df['TOTAL MP'])
```

En efecto, según el test Dickey-Fuller (creo) la serie no es estacionaria.

ESTO ESTÁ MAL, NO PODEMOS PONERNOS A ESTIMAR LOS ÓRDENES DE SARIMA ANTES DE HACER LA SERIE ESTACIONARIA: El hecho de que los valores del ACF van decayendo exponencialmente junto con el lag significativo del PACF sugieren un modelo AR(2). Vemos una clara estacionalidad tras 12 meses. Como hay lags significativos a medida que se va acercando al lag 12, se sugiere que también puede haber un MA de orden 1, 2 o 3 (no lo sé).

Creo que lo que hay que hacer es diferenciar la serie y luego ya sacar los pdqs.

Vemos que tras una diferenciación de primer orden la serie ya sí que es estacionaria, por lo que basándonos en el ACF y el PACF de la serie diferenciada, podemos empezar a construir el modelo SARIMA.

menor bic -> mejor modelo

```python
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Cargar los datos
# Asegúrate de que 'df' es el DataFrame con la columna 'TOTAL MP'
df = monthly_df[['YEAR', 'MONTH', 'TOTAL MP']]
df['DATE'] = pd.to_datetime(df[['YEAR', 'MONTH']].assign(day=1))
df.set_index('DATE', inplace=True)
df = df[['TOTAL MP']]# Verificar estacionariedad
check_stationarity(df['TOTAL MP'])

# Si no es estacionaria, aplicamos diferenciación (primera diferencia)

##esta primera diferenciación se aplica para eliminar la tendencia
df['TOTAL MP Diff'] = df['TOTAL MP'].diff().dropna()
check_stationarity(df['TOTAL MP Diff'].dropna())

# Descomposición para ver tendencia y estacionalidad
result = seasonal_decompose(df['TOTAL MP'], model='additive', period=12)
result.plot()
plt.show()

# Ajustar modelo SARIMA(p,d,q)(P,D,Q,s)
model5 = SARIMAX(df['TOTAL MP'], order=(0,1,0), seasonal_order=(1,0,1,12))
results5 = model5.fit()

# Resumen del modelo
print(results5.summary())

# Diagnóstico de residuales
results5.plot_diagnostics(figsize=(12, 6))
plt.show()

# Pronóstico a 12 meses
forecast = results5.get_forecast(steps=12)
forecast_index = pd.date_range(df.index[-1], periods=12, freq='M')
forecast_values = forecast.predicted_mean

# Graficar pronóstico
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['TOTAL MP'], label='Observado')
plt.plot(forecast_index, forecast_values, label='Pronóstico', linestyle='dashed')
plt.legend()
plt.xlabel('Fecha')
plt.ylabel('Total MP')
plt.title('Pronóstico SARIMA')
plt.show()
```

^^^este es el mejor modelo so far^^^^

### Análisis de los gráficos


*   El primero muestra los residuos estandarizados a lo largo del tiempo. Deberían comportarse como ruido blanco, es decir, fluctuar aleatoriamente alrededor de 0, como es el caso. El problema es que hay algunos picos significativos que sugieren posibles eventos inusuales.
*   El histograma con densidad estimada muestra el histograma de los residuos junto con la densidad estimada (KDE) y la densidad normal estándar (N(0,1)). En este caso, la densidad estimada se desvía significativamente de la normal estándar. Se observa un sesgo negativo y colas pesadas, lo que confirma la no normalidad de los residuos.
*   EL gráfico Q-Q compara los cuantiles de los residuos con los cuantiles de una distribución normal estándar. En este caso, los puntos se desvían significativamente de la línea roja, especialmente en las colas, lo que confirma aún más la no normalidad de los residuos.



### CONCLUSIONES DEL MODELO SARIMA(0,1,0)X(1,0,1)12:


*   El coeficiente AR estacional de orden 1 (AR(12)) es 0.9994, lo cual es muy cercano a 1. Esto sugiere una fuerte estacionalidad autorregresiva.
*   El coeficiente MA estacional de orden 1 (MA(12)) es -0.9796, lo cual también es cercano a -1. Esto indica una fuerte componente de media móvil estacional.
*   La varianza del error es 5.814e+05.
*   Tanto los coeficientes AR como MA estacionales son altamente significativos (P < 0.001).
*AIC = 2793.742
BIC = 2803.167
Estos valores pueden usarse para comparar este modelo con otros modelos alternativos
*Ljung-Box (L1): El valor de Q es 0.02 con una probabilidad de 0.90. Esto sugiere que no hay autocorrelación significativa en los residuos, lo cual es bueno.
*Jarque-Bera (JB): El valor de JB es 255.18 con una probabilidad de 0.00. Esto indica que los residuos no están normalmente distribuidos, lo cual es una preocupación. El Skew (-1.19) y la Kurtosis (8.49) también apoyan esta conclusión.
*Heteroskedasticity (H): El valor de H es 2.70 con una probabilidad de 0.00. Esto sugiere que hay heterocedasticidad en los residuos (la varianza de los residuos no es constante a lo largo del tiempo), lo cual también es una preocupación.

#### Conclusiones del modelo:
* Fortalezas: captura la fuerte estacionalidad de los datos, como indican los coeficientes cercanos a 1 para los AR y MA estacionales y no hay autocorrelación significativa entre los residuos.
* Debilidades: los residuos no están normalmente distribuidos y muestran heterocedasticidad y los coeficientes están muy cercanos a 1, lo cual puede sugerir que la serie temporal es casi estacionalmente no estacionaria (la estacionalidad en sí misma no es constante, sino que cambia con el tiempo)
